```python
import os

folder_path = "images/photos/"
gallery_html = ""

# 遍历文件夹中的文件
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(folder_path, filename)
        gallery_html += f'<img src="{image_path}">'
# 生成HTML文件
html_content = f"""
<!DOCTYPE html>
<html>
<head>
  <style>
    .gallery {{
      display: flex;
      flex-wrap: wrap;
      justify-content: center;
      align-items: center;
      gap: 20px;
      max-width: 800px;
      margin: 0 auto;
    }}
    .gallery img {{
      width: 200px;
      height: 200px;
      object-fit: cover;
    }}
  </style>
</head>
<body>
  <div class="gallery">
    {gallery_html}
  </div>
</body>
</html>
"""

# 将HTML内容保存到文件
with open("gallery.html", "w") as file:
    file.write(html_content)
```

```python
import os

# 指定图片文件夹路径
folder_path = "images/photos/"

# 获取文件夹中的图片文件名列表
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and f.endswith(('.jpg', '.jpeg', '.png'))]

# 生成HTML代码
html_code = '<div class="gallery">\n'
for i, image_file in enumerate(image_files):
    image_id = "image" + str(i+1)
    image_src = os.path.join(folder_path, image_file)
    html_code += f'  <img id="{image_id}" src="{image_src}">\n'
html_code += "</div>"

# 输出HTML代码
print(html_code)
```

```python
import os
from PIL import Image

def compress_images(input_dir, output_dir, quality):
    # 创建输出文件夹
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 遍历输入文件夹中的所有文件
    for filename in os.listdir(input_dir):
        if filename.endswith('.jpg'):
            # 构建输入和输出文件的路径
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, filename)

            # 打开图片并进行压缩
            with Image.open(input_path) as img:
                img.save(output_path, 'JPEG', quality=quality)

            print(f'压缩成功: {output_path}')

# 输入文件夹路径
input_dir = 'images/photos/'

# 输出文件夹路径
output_dir = 'images/compressed/'

# 压缩质量（0-100），值越低压缩率越高
quality = 1

# 调用函数进行压缩
compress_images(input_dir, output_dir, quality)
```
